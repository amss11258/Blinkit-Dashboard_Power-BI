# Step 8: outliers from the dataset according to the columns.

```python
# Calculate Z-scores for sale_price
mean_price = df["sale_price"].mean()
std_price = df["sale_price"].std()
z_scores = (df["sale_price"] - mean_price) / std_price

# Mark outliers (absolute z-score > 3)
outliers = np.abs(z_scores) > 3

# Replace outliers with mean
df.loc[outliers, "sale_price"] = mean_price

# Check if outliers remain
print("Outliers before:", outliers.sum())
print("Outliers after:", ((df["sale_price"] - mean_price) / std_price).abs().gt(3).sum())
```

# Step 9: plot or visualizations

```python
# Histogram for Distribution of sales prices
plt.figure(figsize=(9.29,5))
sns.histplot(df['sale_price'], bins=30, kde=True)
plt.title("Distribution of Sale Prices")
plt.show()
```

```python
# Pie chart for categories
category_counts = df["category"].value_counts().head(10)

plt.figure(figsize=(8,8))   # width=8, height=8 (square looks better for pie)
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Top 10 Product Category Share")
plt.show()
```

```python
# Bar Chart Top 10 brands by count
top_brands = df["brand"].value_counts().head(10)
plt.figure(figsize=(9.76,4))
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.xticks(rotation=45)
plt.title("Top 10 Brands by Count")
plt.show()
```
